#### Change train and test labels into one-hot vectors

```
trainY=tf.keras.utils.to_categorical(YTraindata, num_classes=10)
valY=tf.keras.utils.to_categorical(YValdata, num_classes=10)
testy=tf.keras.utils.to_categorical(Y_test, num_classes=10)
# Changing the train and test labels to one-hot vectors
```

#### Build the Graph

#### Initialize model, reshape & normalize data

```
#Initialize Sequential model
model = tf.keras.models.Sequential()

#Reshape data from 2D to 1D -> 28x28 to 784
model.add(tf.keras.layers.Reshape((1024,),input_shape=(32,32,)))

# adding batch normalization
model.add(tf.keras.layers.BatchNormalization())

```

#### Add two fully connected layers with 200 and 100 neurons respectively with `relu` activations. Add a dropout layer with `p=0.25`

```
#Add the first layer
model.add(tf.keras.layers.Dense(200, activation='relu'))
model.add(tf.keras.layers.BatchNormalization())
# Adding dropout to first layer
model.add(tf.keras.layers.Dropout(0.25))
#Add the second layer
model.add(tf.keras.layers.Dense(100, activation='relu'))
#Adding dropout to second layer
model.add(tf.keras.layers.Dropout(0.25))
#Output layer
model.add(tf.keras.layers.Dense(10, activation='softmax', name='Output'))
```

### Add the output layer with a fully connected layer with 10 neurons with `softmax` activation. Use `categorical_crossentropy` loss and `adam` optimizer and train the network. And, report the final validation.

```
# Compiling the model with optimizer as 'Adam' and loss as categorical_crossentropy
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy']) 
```

```
#Train the model
model.fit(Xtrainset,trainY,
          validation_data=(XValset,valY),
          epochs=20,
          batch_size=32,
          shuffle="batch")
```
